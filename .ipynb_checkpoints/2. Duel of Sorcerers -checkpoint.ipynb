```python
print("Part 1:")

#Data Structure Treatments
POWER = {'Fireball': 50, 'Lightning bolt': 40, 'Magic arrow': 10, 'Black Tentacles': 25, 'Contagion': 45}

gandalf = ['Fireball', 'Lightning bolt', 'Lightning bolt', 'Magic arrow', 'Fireball', 'Magic arrow', 'Lightning bolt', 'Fireball', 'Magic arrow', 'Fireball']

gandalf_spells = []

saruman = ['Contagion', 'Contagion', 'Black Tentacles', 'Fireball', 'Black Tentacles', 'Lightning bolt', 'Magic arrow', 'Contagion', 'Magic arrow', 'Magic arrow']

saruman_spells = []

x = 0
y = 0

#Spell Assignment
for i in gandalf:
  if gandalf[x] in POWER.keys():
    gandalf_spells.append(POWER[gandalf[x]])    
  if len(gandalf_spells) < len(gandalf):  
    x += 1
  

print('\nGandalf_Spells =', gandalf_spells)

for i in saruman:
  if saruman[y] in POWER.keys():
    saruman_spells.append(POWER[saruman[y]]) 
  if len(saruman_spells) < len(saruman):   
    y += 1

print('\nSaruman_Spells =', saruman_spells)

#Execution of Spell Clashes
battle_results = []
a = 0

for spells in gandalf_spells:
  battle = gandalf_spells[a] - saruman_spells[a]
  battle_results.append(battle)
  a += 1

print('\nbattle_results =', battle_results)

#Check for 3 Wins
gandalf_victories = []
saruman_victories = []
b = 0

for results in battle_results:
  if battle_results[b] > 0:
    gandalf_victories.append('win')
  elif battle_results[b] < 0:
    saruman_victories.append('win')
  
  b += 1
  
  if len(gandalf_victories) == 3:
    print('\nGandlaf is the winner!')
    break
  elif len(saruman_victories) == 3:
    ('\nSaruman is the winner!')
    break

#Statistics Calculations
import statistics

gandalf_stats = []
saruman_stats = []
g = 0
s = 0

for metric in gandalf_spells:
  gandalf_stats.append(gandalf_spells[g])
  g += 1

for metric in saruman_spells:
  saruman_stats.append(saruman_spells[s])
  s += 1

#Gandalf Stats 
gandalf_average = statistics.mean(gandalf_stats)
gandalf_standardev = statistics.stdev(gandalf_stats)
print("\nGandalaf's spells' average =", gandalf_average)
print("\nGandalaf's spells' standard deviation =", round(gandalf_standardev, 2))

#Gandalf Stats 
saruman_average = statistics.mean(saruman_stats)
saruman_standardev = statistics.stdev(saruman_stats)
print("\nSaruman's spells' average =", saruman_average)
print("\nSaruman's spells' standard deviation =", round(saruman_standardev, 2))
```
